```python
# Import the Image,ImageOps, Imagefilter, ImageEnhance modules from the PIL library
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image, ImageOps, ImageFilter, ImageEnhance
```

```python
# Opens the image file
img = Image.open('input.jpg')
```

```python
# Applying image filter using detail
img_detail = img.filter(ImageFilter.DETAIL)

# shows image in image viewer 
img_detail.show()
plt.imshow(img)
```

```python
# Applying image filter using sharpen
img_sharpen = img.filter(ImageFilter.SHARPEN)
img_sharpen.show()
plt.imshow(img)
```

```python
# Applying image filter using MedianFilter
img_median = img.filter(ImageFilter.MedianFilter)
img_median.show()
plt.imshow(img)
```

```python
# Applying image filter using UnsharpMask
img_median = img.filter(ImageFilter.UnsharpMask(5.0, 200, 20))
img_median.show()
plt.imshow(img)
```

```python
# The following Histogram equalization from ImageOps
img_histogram = Image.open('input.jpg')
img_histogram = ImageOps.equalize(img_histogram)
img_histogram = img_histogram.save("img_histogram.jpg")

img_histogram = Image.open('img_histogram.jpg')
img_histogram.show()
plt.imshow(img_histogram)
```
